```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('multiple_linear_regression_dataset.csv')
df.columns = ['X1','X2','Y']
df
```

```python
df.info()
```

```python
lr_multi = LinearRegression()
lr_multi.fit(df[['X1','X2']],df['Y'])
pred = lr_multi.predict(df[['X1','X2']])
r2 = r2_score(df['Y'],pred)
print('r2_score = ',r2)
print('coefficients = ',lr_multi.coef_)
print('intercept = ',lr_multi.intercept_)
```

```python
#X1~X2
LR12 = LinearRegression()
LR12.fit(df[['X2']],df['X1'])
pred = LR12.predict(df[['X2']])
X12 = df['X1'] - pred
df['X12'] = round(X12,2)
```

```python
#X2 ~X1
LR21 = LinearRegression()
LR21.fit(df[['X1']],df['X2'])
pred = LR21.predict(df[['X1']])
X21 = df['X2'] - pred
df['X21'] = round(X21,2)
```

```python
df
```

```python
# X1 ~ y
lrX1Y = LinearRegression()
lrX1Y.fit(df[['X1']],df['Y'])
pred = lrX1Y.predict(df[['X1']])
Y1 = df['Y'] - pred
RX1 = r2_score(df['Y'],pred)
df['Y1'] = round(Y1,2)
```

```python
# X2 ~ y
lrX2Y = LinearRegression()
lrX2Y.fit(df[['X2']],df['Y'])
pred = lrX2Y.predict(df[['X2']])
Y2 = df['Y'] - pred
RX2 = r2_score(df['Y'],pred)
df['Y2'] = round(Y2,2)
```

```python
df
```

```python
# X12 ~ y
lrX12Y = LinearRegression()
lrX12Y.fit(df[['X12']],df['Y'])
b1 = lrX12Y.coef_[0]
b1_scaled = (b1/df['Y'].std())*df['X12'].std()
```

```python
# X21 ~ y
lrX21Y = LinearRegression()
lrX21Y.fit(df[['X21']],df['Y'])
b2 = lrX21Y.coef_[0]
b2_scaled = (b2/df['Y'].std())*df['X21'].std()
```

```python
R = df.corr()['Y']
```

```python
sr1 = R['X12']
sr2 = R['X21']
```

```python
pr1 = df.corr()['Y2']['X12']
pr2 = df.corr()['Y1']['X21']
```

```python
LR_X1_X2 = LinearRegression()
LR_X1_X2.fit(df[['X1','X2']],df['Y'])
pred = LR_X1_X2.predict(df[['X1','X2']])
r2_X1_X2 = r2_score(df['Y'],pred)
```

```python
print('sr1 = ',round(sr1,2)) # semi partial correlation of X1
print('pr1 = ',round(pr1,2)) #partial correlation of X1
print('sr2 = ',round(sr2,2)) # semi partial correlation of X2
print('pr2 = ',round(pr2,2)) #partial correlation of X2
print('b1 scaled',b1_scaled)
print('b2 scaled',b2_scaled)
print()
print('r2_score y~X1', RX1)
print('r2_score y~X2', RX2)
print('r2_score y~[X1,X2]', r2_X1_X2)
```

```python
print('Contribution of X1 to variance explained in Y is ',sr1**2)
print('Contribution of X2 to variance explained in Y is ',sr2**2)
```
